***Exportando o Dataset***

```python
url = 'https://raw.githubusercontent.com/GuilhermePiresOliveira/Datasets/refs/heads/main/StudentPerformanceFactors.csv'
```

***Importando as Bibliotecas***

```python
import numpy as np # álgebra linear
import pandas as pd # processamento de dados, E/S do arquivo CSV (por exemplo, pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

***Lendo o Dataset***

```python
df = pd.read_csv(url)
```

***Definindo começa pelo índice 1***

```python
df.index = df.index + 1
```

***Visualizado o Dataset***

```python
df.head()
```

***Traduzindo as colunas***

```python
colunas_traduzidas = {
    'Hours_Studied': 'Horas_Estudadas',
    'Attendance': 'Frequencia',
    'Parental_Involvement': 'Envolvimento_Parental',
    'Access_to_Resources': 'Acesso_a_Recursos',
    'Extracurricular_Activities': 'Atividades_Extracurriculares',
    'Sleep_Hours': 'Horas_de_Sono',
    'Previous_Scores': 'Notas_Anteriores',
    'Motivation_Level': 'Nivel_de_Motivacao',
    'Internet_Access': 'Acesso_a_Internet',
    'Tutoring_Sessions': 'Sessoes_de_Tutoria',
    'Family_Income': 'Renda_Familiar',
    'Teacher_Quality': 'Qualidade_do_Professor',
    'School_Type': 'Tipo_de_Escola',
    'Peer_Influence': 'Influencia_dos_Colegas',
    'Physical_Activity': 'Atividade_Fisica',
    'Learning_Disabilities': 'Deficiencias_de_Aprendizagem',
    'Parental_Education_Level': 'Nivel_de_Educacao_dos_Pais',
    'Distance_from_Home': 'Distancia_de_Casa',
    'Gender': 'Genero',
    'Exam_Score': 'Nota_no_Exame'
}
```

```python
df.rename(columns=colunas_traduzidas, inplace=True)
```

```python
df.head()
```

***Análise de desempenho das notas***

```python
nota = df['Nota_no_Exame']
plt.figure(figsize=(8, 6))
plt.hist(nota, bins=10, edgecolor='black')
plt.title('Distribuição das pontuações dos exames')
plt.xlabel('Nota_no_Exame')
plt.ylabel('Numero de Estudantes')
plt.grid(True)
plt.show()
```




*A maioria das notas se concentra entre 60 e 75 pontos.*

*Poucos estudantes tiveram notas abaixo de 60 ou acima de 80, o que mostra que as pontuações estão bastante concentradas em torno de uma média.*

**Isso sugere que, no geral, a turma teve um desempenho médio, com poucas exceções extremas tanto para notas muito baixas quanto muito altas.**
